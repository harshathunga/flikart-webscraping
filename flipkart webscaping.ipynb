```python
from bs4 import BeautifulSoup
import requests
import openpyxl
import pandas as pd

excel = openpyxl.Workbook()
#  excel.sheetnames is to check howmany sheets are there in the workbook
print(excel.sheetnames)
sheet = excel.active
sheet.title = 'flipkart cam data'
print(excel.sheetnames)
sheet.append(['head'])
```

```python
p = 1
p_name=[]
p_price=[]
p_descip=[]
for i in range(1,3+1):
    p = i
    url = 'https://www.flipkart.com/search?q=mobile+5g&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_3_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobile+5g%7CMobiles&requestId=82f40fc3-1018-4826-bbd5-91826175649c&as-backfill=on&page='+str(p)
    print(url)
    page = requests.get(url)
    print(page)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    # print(soup)
    but2= soup.find_all('div',class_ ='_1AtVbE col-12-12')
    
    names = soup.find_all('div', class_ = "_4rR01T")
    
    for i in names:
        nam = i.text
        
        p_name.append(nam)
        # sheet.append(p_name)
    
    price = soup.find_all('div', class_='_30jeq3 _1_WHN1')
    
    for i in price:
        pr = i.text
        # sheet.append([pr])
        p_price.append(pr)
    
    des =soup.find_all('ul', class_ = '_1xgFaf')
    for i in des:
        desv = i.text
        p_descip.append(desv)


print(len(p_name))
print(len(p_price))
print(len(p_descip))
    # page = requests.get(url)
    # print(page)
    # soup = BeautifulSoup(page.text, 'html.parser')
```

```python
data = pd.DataFrame({"p_name": p_name, "p_price": p_price, "p_description": p_descip})
data

```

```python
data.to_excel("webscrape flipkart data.xlsx")
```
